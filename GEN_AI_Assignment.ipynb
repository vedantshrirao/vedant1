```
df = pd.read_csv("/content/data1.csv")
```

```
symptoms = {
    'Symptom1': {'Fever': ['Mild', 'Low', 'High']},
    'Symptom2': {'Cough': ['Mild', 'Low', 'High']},
    'Symptom3': {'Cold': ['Mild', 'Low', 'High']}
}

print(symptoms)
```

```
# Extract unique symptoms from the "Other_Symptoms" column
other_symptoms = df['Other_Symptoms'].str.split(',').explode().str.strip().unique()

# Prepare a list to track losses (symptoms not found in the dictionary)
loss = []

# Check each unique symptom in "Other_Symptoms" and compare with the dictionary
for symptom in other_symptoms:
    found = False
    # Check if symptom exists in any of the symptom groups in the dictionary
    for symptom_group, symptom_info in symptoms.items():
        if symptom in symptom_info:
            found = True
            break
    # If not found, add it to the dictionary and log it as loss
    if not found:
        loss.append(symptom)
        new_symptom_group = f'Symptom{len(symptoms) + 1}'
        # symptoms[new_symptom_group] = {symptom: ['Mild', 'Low', 'High']}
        symptoms[new_symptom_group] = {symptom:None}

# Print the updated symptoms dictionary
print("Updated Symptoms Dictionary:")
for symptom_group, symptom_info in symptoms.items():
    print(f"{symptom_group}: {symptom_info}")

# Print the loss
print("\nLoss (new symptoms added to the dictionary):")
print(loss)
```

```
# Calculate and print loss as a number
total_columns = len(symptoms)
loss_number = len(loss) / total_columns if total_columns > 0 else 0
print("\nLoss as a number:", loss_number)
```

Step 4: Parser is nearly able to
1. Read any format of data csv,tsv,json,xml  
2. Reads and enhances dictionary
3. In next pass reads the data and prints as per the available colums in
   dictonary
4. Use a facility to dump the dictionary as file  
5. Allow manual editing and reparsing of dictionary

```
import pandas as pd
import json
import xml.etree.ElementTree as ET

class SymptomParser:
    def __init__(self, dictionary_file='symptoms_dict.json'):
        self.symptom_dictionary = {}
        self.dictionary_file = dictionary_file
        self.load_dictionary()

    # Method to read data from various formats
    def load_data(self, file_path):
        if file_path.endswith('.csv'):
            return pd.read_csv(file_path)
        elif file_path.endswith('.tsv'):
            return pd.read_csv(file_path, delimiter='\t')
        elif file_path.endswith('.json'):
            return pd.read_json(file_path)
        elif file_path.endswith('.xml'):
            return self.convert_xml_to_dataframe(file_path)
        else:
            raise ValueError("Unsupported file format")

    # Custom XML parser to convert data to DataFrame
    @staticmethod
    def convert_xml_to_dataframe(file_path):
        tree = ET.parse(file_path)
        root = tree.getroot()
        records = []
        for patient in root.findall('./Patient'):
            record = {}
            for child in patient:
                record[child.tag] = child.text
            records.append(record)
        return pd.DataFrame(records)

    # Method to enrich the dictionary by adding symptoms missing from the dataset
    def enrich_dictionary(self, dataset, symptom_column='Other_Symptoms'):
        for symptom_list in dataset[symptom_column].dropna():
            for symptom in symptom_list.split(','):
                symptom = symptom.strip()  # Trim whitespace

                # Check if the symptom is already present in the dictionary
                exists = any(symptom in group for group in self.symptom_dictionary.values())

                if not exists:
                    symptom_key = f'Symptom{len(self.symptom_dictionary) + 1}'
                    self.symptom_dictionary[symptom_key] = {symptom: ['Mild', 'Low', 'High']}

        return self.symptom_dictionary

    # Method to filter and display data based on the available symptoms in the dictionary
    def display_filtered_data(self, dataset, symptom_column='Other_Symptoms'):
        available_symptoms = {symptom for group in self.symptom_dictionary.values() for symptom in group}
        filtered = dataset[dataset[symptom_column].apply(lambda x: any(symptom in x for symptom in available_symptoms) if pd.notnull(x) else False)]
        print("Filtered Data Based on Available Symptoms in Dictionary:")
        print(filtered)

    # Method to save the dictionary to a file
    def save_dictionary(self):
        with open(self.dictionary_file, 'w') as file:
            json.dump(self.symptom_dictionary, file, indent=4)
        print(f"Dictionary saved to {self.dictionary_file}")

    # Method to load the dictionary from a file
    def load_dictionary(self):
        try:
            with open(self.dictionary_file, 'r') as file:
                self.symptom_dictionary = json.load(file)
        except FileNotFoundError:
            print(f"No existing dictionary found at {self.dictionary_file}. Starting fresh.")
            self.symptom_dictionary = {}

    # Method for manual updates to the dictionary
    def update_dictionary_manually(self):
        symptom = input("Enter the symptom you wish to add: ")
        levels = input("Specify severity levels (comma-separated, e.g., Mild, Low, High): ")
        severity_levels = [level.strip() for level in levels.split(',')]

        if any(symptom in group for group in self.symptom_dictionary.values()):
            print(f"The symptom '{symptom}' already exists in the dictionary.")
        else:
            symptom_key = f'Symptom{len(self.symptom_dictionary) + 1}'
            self.symptom_dictionary[symptom_key] = {symptom: severity_levels}
            print(f"Added '{symptom}' with levels {severity_levels} to the dictionary.")
        self.save_dictionary()

    # Main method to handle the data processing
    def handle_data(self, file_path, symptom_column='Other_Symptoms'):
        # Step 1: Load data
        data = self.load_data(file_path)

        # Step 2: Display data according to available dictionary entries
        self.display_filtered_data(data, symptom_column=symptom_column)

        # Step 3: Enrich the dictionary with new symptoms from the dataset
        self.enrich_dictionary(data, symptom_column=symptom_column)

        # Step 4: Re-display the filtered data based on the updated dictionary
        self.display_filtered_data(data, symptom_column=symptom_column)

        # Step 5: Save the updated dictionary to a file for future reference
        self.save_dictionary()

# Example usage:
if __name__ == "__main__":
    # Instantiate the parser
    parser = SymptomParser(dictionary_file='symptoms_dict.json')

    # Process a data file (supports CSV, TSV, JSON, XML formats)
    file_path = '/content/data1.csv'  # Replace with your actual file path
    parser.handle_data(file_path, symptom_column='Other_Symptoms')

    if input("Do you want to manually update the dictionary? [Y/N] ").strip().upper() == "Y":
        parser.update_dictionary_manually()
```
